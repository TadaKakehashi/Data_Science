### Pandas - DataFrame and Series
Pandas is a powerful data manipulation library, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. 
A Series is a one-dimensional array-like object, while DataFrame is a two dimensional, size-mutable and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
## Series: Its a 1D array like object which can hold any data type
import pandas as pd

data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
## Create a Series from a Dictionary
data = {'a':1,'b':2,'c':3} #keys becomes default indexing
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30,40,50]
index = ['a','b','c','d','e']

pd.Series(data,index=index) # creating our own series with index
```

```python
## DataFrame
## Creating a dataframe from a dictionary of list

data ={
    'Name': ['Tada','Ray','Aryan'],
    'Age':[24,23,22],
    'City':['NewYork','Tokyo','Delhi']
}

df = pd.DataFrame(data)
print(df)
print()
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Creating a DataFrame from a list of dictionaries 


import pandas as pd

data = [
    {'Name':'Tada','Age':21,'City':'Bhubaneswar','Salary':45000,'Department':'Data Science'},
    {'Name':'Ray','Age':23,'City':'Cuttack','Salary':52000,'Department':'AI'},
    {'Name':'Aryan','Age':22,'City':'Hanspal','Salary':48000,'Department':'Backend'},
    {'Name':'Neha','Age':24,'City':'Delhi','Salary':60000,'Department':'ML'},
    {'Name':'Riya','Age':25,'City':'Mumbai','Salary':65000,'Department':'Analytics'},
    {'Name':'Karan','Age':26,'City':'Bangalore','Salary':70000,'Department':'DevOps'}
]

df = pd.DataFrame(data)
print(df)
```
