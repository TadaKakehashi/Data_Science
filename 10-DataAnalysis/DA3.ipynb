```python
import pandas as pd
import numpy as np

np.random.seed(42)

rows = 120

data = pd.DataFrame({
    "ID": range(1, rows + 1),
    "Name": [f"Person_{i}" for i in range(1, rows + 1)],
    "Age": np.random.randint(18, 60, rows),
    "City": np.random.choice(["Chennai", "Mumbai", "Delhi", "Bangalore", "Hyderabad", "Kolkata"], rows),
    "Salary": np.random.randint(25000, 120000, rows),
    "Experience_Years": np.random.randint(0, 20, rows),
    "Department": np.random.choice(["Data", "HR", "Finance", "Engineering", "Sales"], rows),
    "Performance_Score": np.round(np.random.uniform(1, 5, rows), 2)
})

# Add missing values randomly
data.loc[np.random.choice(data.index, 10), "Age"] = np.nan
data.loc[np.random.choice(data.index, 8), "City"] = np.nan
data.loc[np.random.choice(data.index, 12), "Salary"] = np.nan
data.loc[np.random.choice(data.index, 6), "Performance_Score"] = np.nan

# Add some duplicate rows
data = pd.concat([data, data.iloc[5:10]])

# Shuffle dataset
data = data.sample(frac=1).reset_index(drop=True)

data.to_csv("messy_data.csv", index=False)

print("Messy dataset created!")
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "EmpID": [1,2,3,4,5],
    "Name": ["Alice","Bob","Charlie","David","Eva"],
    "Department": ["Data","HR","Engineering","Data","Finance"]
})

print(df1)
```

```python
df2 = pd.DataFrame({
    "EmpID": [1,2,3,6],
    "Salary": [50000,45000,70000,60000],
    "City": ["Chennai","Mumbai","Delhi","Bangalore"]
})

print(df2)
```

```python
## Merge DataFrame on the 'Key Columns'
pd.merge(df1,df2,on="EmpID",how="inner")
```

```python
pd.merge(df1,df2,on="EmpID",how="outer")
```

```python
pd.merge(df1,df2,on="EmpID",how="left")
```

```python
pd.merge(df1,df2,on="EmpID",how="right")
```
