## Data Visualization with Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of codes. 

```python
import seaborn as sns
print(sns.__version__)
```

```python
import seaborn as sns
import pandas as pd

data = {
    "total_bill": [16.99, 10.34, 21.01, 23.68, 24.59, 25.29, 8.77, 26.88, 15.04, 14.78],
    "tip": [1.01, 1.66, 3.50, 3.31, 3.61, 4.71, 2.00, 3.12, 1.96, 3.23],
    "sex": ["Female","Male","Male","Male","Female","Male","Male","Male","Female","Male"],
    "smoker": ["No","No","No","No","No","No","No","Yes","Yes","Yes"],
    "day": ["Sun","Sun","Sun","Sun","Sun","Sat","Sat","Sat","Sat","Sat"],
    "time": ["Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner"],
    "size": [2,3,3,2,4,4,2,4,2,2]
}

tips = pd.DataFrame(data)

print(tips)
```

```python
# Basic Plotting with Seaborn

tips = sns.load_dataset('tips')
print(tips)
```

```python
## Creating a Scatter plot

import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)

plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot
sns.lineplot(x='size',y='total_bill',data = tips)
plt.title("Line Plot of Total Bill")
```

```python
## Categorical plots
## Bar Plot
sns.barplot(x='day',y='total_bill',data = tips)
plt.title('Bar plot of Total Bill of Day')
```

```python
### Box Plot
sns.boxplot(x="day",y='total_bill',data =tips)
plt.show()
```

```python
## Violin Plot
sns.violinplot(x='day',y='total_bill',data = tips,color='red')
plt.show()
```

```python
## Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.show()
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True,color='green') #fill = shade
```

```python
# Pair plot -> relation between two different variables
sns.pairplot(tips)
```

```python
#HeatMap
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
