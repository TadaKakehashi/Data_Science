### Importing Modules in Python: Modules and Packages
It helps in organizing and reuse code. 

```python
import math
math.sqrt(16)
```

```python
from math import sqrt, pi 
print(sqrt(16))
print(pi)
```

```python
import numpy as np #np -> alias for numpy
np.array([1,2,3,4,5])
```

```python
from math import * # importing all the functions defined in math
sqrt(16)
```

```python
from package.maths import addition
addition(1,2)
```

```python
from package import maths
maths.addition(2,3)
```
