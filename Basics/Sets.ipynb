## Sets

Sets are used to store unique items in unordered manner and they do not allow any duplicate elements. 

```python
##create a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set((1,2,3,4)) #typecasting into a set
print(my_set)
```

```python
my_set = set([1,2,3,4,4,4,5])
print(my_set)
```

```python
## Basic set operations
## Adding and Removing elements

my_set.add(3)
print(my_set)
my_set.add(7)
print(my_set)

my_set.remove(3)
print(my_set)
```

```python
my_set.discard(3)
print(my_set)
```

```python
## pop method

removed_element = my_set.pop() #first in first out hence we get the first element
print(removed_element)
print(my_set)
```

```python
my_set.clear()
print(my_set)
```

```python
## Set Membership Test
my_set = {1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operations
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

### Union
union_set = set1.union(set2)
print(union_set)

### Intersection -> common elements
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2) #common elements got updated in set1
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
```

```python
## Difference

differ_set = set1.difference(set2)
print(differ_set)

set1.difference_update(set2)
print(set1)
```

```python
## Symmetric Difference -> common elements will be removed
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

sym_set = set1.symmetric_difference(set2)
print(sym_set)
```

```python
## Sets Methods
set1 = {1,2,3,4,5}
set2 = {3,4,5}

## is subset -> all the elements present in set1 is available in set2 or not

print(set1.issubset(set2))

## is superset -> whether it has all the elements of set 2
print(set1.issuperset(set2))
```

```python
lst = [1,2,3,4,5,6,6,6,7]
print(set(lst))
```

```python
### Counting Unique words in text

text = "Hello Tada how are u u"
words = text.split()

## Converting list of words into set

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
