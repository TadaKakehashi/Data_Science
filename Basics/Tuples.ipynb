## Tuples are ordered collections of items that are immutable 

```python
## creating a tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6]) #list converted into a tuple, and vice versa (list(tuple))
print(numbers)
```

```python
mixed_tuple = (1,2,"Taa")
print(mixed_tuple)
```

```python
## Accessing Tuple Elements
numbers[0] 
print(numbers[-1])

print(numbers[0:4])
print(numbers[::-1])
```

```python
## Tuple Operations!

Concat_tup = numbers + mixed_tuple
print(Concat_tup)
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```
