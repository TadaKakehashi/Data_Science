### Generators
Generators are a simple way to create iterators. They use the "yield" keyword to produce a series of value lazily which means they generate values on the fly and do not store them in memory.

```python
def square(n):
    for i in range(3):
        yield i**2
```

```python
print(square(3))
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
a
```

```python
next(a) #becomes a generator and we use next 
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
print(gen)
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```
