```python
#function copy
def welcome():
    return "Welcome"

welcome()
```

```python
wel = welcome
print(wel())

del welcome #deleting the function
print(wel())
```

```python
## Closures -> Function inside a function

def main_welcome():
    msg = "Welcome"
    def sub_welcome():
        print("Welcome Tada!")
        print(msg)
        print("Good Morning!")
    return sub_welcome()
```

```python
main_welcome()
```

```python
def main_welcome(func):

    def sub_welcome():
        print("Welcome Tada!")
        func()
        print("Good Morning!")
    return sub_welcome()
```
