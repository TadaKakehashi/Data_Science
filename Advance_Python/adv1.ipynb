### Iterators
Iterators are advance concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list = [1,2,3,4,5]
for i in my_list:
    print(i)
```

```python
## Iterator
Iterator = iter(my_list)
print(type(Iterator))
```

```python
Iterator ##It uses lazy loading technique
```

```python
## Iterate through all the element
next(Iterator) #accessing 1 value here, and we move one step ahead a time. As it reaches last element it will share as an error message
```

```python
my_list = [1,2]
iterator = iter(my_list)

try:
    print(next(iterator))
    print(next(iterator))
    print(next(iterator))  # No element left
except StopIteration:
    print("There are no elements in the iterator")
```

```python
## String Iterator
my_string = "Hello"
string_iterator = iter(my_string)

print(next(string_iterator)) 
print(next(string_iterator))
```
