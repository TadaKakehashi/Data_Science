# Map()
The map() function applies a given function to all items in an input list(or any other iterable) and returns a map object(an iterator). This is particularly useful for transforming data in a list comprehensively.


```python
lst = [1,2,3,4,6,7,8,9]
sq = list(map(lambda x:x*x, lst))
print(sq)
```

```python
## Using map to convert a list of strings to integers
# Use map to convert strings to integers

lst = ['1','2','3','4']
str_to_int = list(map(int, lst))
print(str_to_int)
```

```python
lst1 = [1,2,3]
lst2 = [3,4,5]

combined = list(map(lambda x,y:x*y, lst1,lst2))
print(combined)
```

```python
words = ['apple','banana','cherry']
upper_word = list(map(str.upper,words))

print(upper_word)
```

```python
#Dict to list

def get_name(person):
    return person['name']

people =[
    {'name':'aryan','age':10},
    {'name':'ray','age':12},
    {'name':'tada','age':15},
]

list(map(get_name,people))
```

The map() function is a powerful tool for applying transformations to iterable data structures. It can be used with regular functions, lambda functions, and even multiple iterables, providing a versatile approach to data processing in Python.
