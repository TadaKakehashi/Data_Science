Exception are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution.
Common exceptions include:->
    ZeroDivision: Diving by Zero,
    File Not Found Error,
    TypeError,

```python
## Exception handling, -> try ,except block

try:
    a = b
except NameError as ex:
    print(ex)
```

```python
try:
    result = 1/0
except ZeroDivisionError as ex:
    print(ex)
```

```python
try:
    result = 1/0
    a = c
except Exception as ex1:
    print(ex1)
```

```python
try:
    o = p
    total = 1/0
except ZeroDivisionError as exp:
    print(exp)
    print("Exception 1")
except Exception as ex2:
    print(ex2)
    print("Exception 2")
```

```python
try:
    num = int(input("Enter number"))
    result = 10/num
except ValueError:
    print("This is not a valid number!")
except ZeroDivisionError:
    print("Enter Denominator greater than 0")
except Exception as ex:
    print(ex)
```

```python
## TRY , Except, else block
try:
    num = int(input("Enter a number!"))
    result = 10/num
except ValueError as ex1:
    print(ex1)
except ZeroDivisionError as ex2:
    print(ex2)
except Exception as ex3:
    print(ex3)
else:
    print(f"The Result is {result}")
```

```python
## try,except,else and finally blocks
try:
    num = int(input("Enter a number!"))
    result = 10/num
except ValueError as ex1:
    print(ex1)
except ZeroDivisionError as ex2:
    print(ex2)
except Exception as ex3:
    print(ex3)
else:
    print(f"The Result is {result}")
finally:
    print("Execution complete.") #whether error occurs or not it will get executed!
```

```python
## File handling and Exception Handling
try: 
    file = open('example.txt','r')
    content = file.read()
    print(content)

except FileNotFoundError:
    print("File does not exist!")

finally:
    if 'file' in locals() or not file.closed():
        file.close()
        print('File Closed!')
```
