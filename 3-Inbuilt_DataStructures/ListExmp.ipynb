```python
## Example1: Manage a Todo List

to_do_list = ["Study","Clean","Drive"]
print(to_do_list)

## Adding to the todo list
to_do_list.append("Schedule Meeting")
to_do_list.append("Go to Gym")
print(to_do_list)

## Removing particular task after completion

to_do_list.remove("Clean")
print(to_do_list)

if "Study" in to_do_list:
    print("Dont forgot to study!")

print("To do list remaining!")
print()
for task in to_do_list:
    print(f'{task}')
```
