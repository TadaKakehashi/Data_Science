## Dictionaries

They are unordered collections of items. They store data in key-value pairs. keys must be unique and immutable. (eg. Strings, numbers, or tuples) while values can be of any type.

```python
## Creating Dictionaries
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
print(empty_dict)
```

```python
students = {
    "name": "Tada",
    "age": 23,
    "grade": 'A',
}

print(students)
```

```python
students = {
    "name": "Tada",
    "age": 23,
    "grade": 'A',
    "name": "Ray" #recent value will get replaced
} 

print(students)
```

```python
## Accessing Dictionary Elements
students = {
    "name": "Tada",
    "age": 23,
    "grade": 'A',
}

print(students['grade'])
print(students['age'])

## Accessing using get() method
print(students.get('grade'))
print(students.get('last_name'))
print(students.get('last_name',"Not Available")) # default value
```

```python
## Modifying dictionary elements
## Dictionary are mutable, so we can add, update and delete elements.

print(students)
```

```python
students["age"] = 33 ## update value for the key
print(students)
students["address"] = "India" ## adding a new key and value
print(students)

del students['address'] ## delete key and value pair
print(students)
```

```python
## Common Dictionary Methods

# To get all the keys
keys = students.keys()
print(keys)

# TO get values
values = students.values()
print(values)

items = students.items() ## get all key value pairs
print(items)
```

```python
## Shallow Copy
student_copy = students
print(student_copy)
```

```python
students["name"] = "Ray"
print(students)
print(student_copy) 
```

```python
student_copy1 = students.copy() ## shallow copy 
print(student_copy1)
students["name"] = "Tada"
print(students)
```

```python
### Iterating over dictionaries
## Loops to Iterate

## Iterating over keys
for key in students.keys():
    print(key)
```

```python
## Iterating over values
for values in students.values():
    print(values)
```

```python
## Iterate over key value pairs

for key, value in students.items():
    print(f"{key} : {value}")
```

```python
## Nested Dictionaries

students = {
    "student1" : {"name":"Tada","age":15},
    "student2" : {"name":"Ray","age":23},
    "student3" : {"name":"Aryan","age":32}
}

print(students)
```

```python
## Accessing nested dictionary elements

print(students["student1"]["name"])
print(students["student3"]["age"])
```

```python
## Iterating over nested dictionary

for student_id,student_info in students.items():
    print(f'{student_id} : {student_info}')
    for key, value in student_info.items():
        print(f'{key} : {value}')
    print()
```

```python
## Dictionary Comprehension (: is important)

squares = {x:x**2 for x in range(5)}
print(squares ,end =" ")
```

```python
## Conditional Dictionary comprehension

evens = {x:x for x in range(10) if x%2==0}
print(evens)
```

```python
## Merge 2 Dictionaries into one

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":5}

merged_dict = {**dict1,**dict2}
print(merged_dict)
```
