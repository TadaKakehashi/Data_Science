### Encapsulation and Abstraction
These two concepts helps in designing robust, maintainable and reusable code. Encapsulation involves bundling data and methods that operate on the data within a single unit, while abstraction involves hiding complex implementation details and exposing only the necessary features.

#### Encapsulation
It is the concept of wrapping data(variables) and methods(functions) together as a single unit. It restricts direct access to some of the object's components, which means of preventing accidental interference and misuse of the data


```python
### Encapsulation with Getter and Setter Methods
### Public, Private, Protected Variables or Access Modifiers

class Person:
    def __init__(self,name,gender,age):
        self.__name = name  ##private variables
        self._gender = gender## protected variable: cant use it outside of the class but can be accessed from a derived class
        self.age = age ##public variables    

    # Getter for private variable
    def get_name(self):
        return self.__name
    # Setter for private variable
    def set_name(self,new_name):
        self.__name = new_name

class Employee(Person):
    def __init__(self,name,gender,age):
        super().__init__(name,gender,age)

employee = Employee("Ray","Male",21)

print("Public age:",employee.age)
print("Protected gender:",employee._gender)
print("Private name(via getter):",employee.get_name())
```

```python
## Encapsulation with getter and setter 
class Person:
    def __init__(self,name,age):
        self.__name = name
        self.__age = age
    
    #Getter
    def get_name(self):
        return self.__name
    #Setter
    def set_name(self,name):
        self.__name = name
    
    def get_age(self):
        return self.__age
    
    def set_age(self,age):
        self.__age = age
        
        
person = Person("Tada",21)
print(person.get_name())
print(person.get_age())

person.set_name("Ray")
print(person.get_name())

person.set_age(69)
print(person.get_age())
```

## Abstraction
It is a concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
from abc import ABC, abstractmethod

## Abstract base class
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")
    
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started!")


def operate_vehicle(vehicle):
    vehicle.start_engine()

car = Car()
operate_vehicle(car)
```
