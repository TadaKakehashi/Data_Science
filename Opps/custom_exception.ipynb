### Custom exception (Raise and Throw an exception)

```python
class Error(Exception):
    pass

class dobException(Error): 
    pass
```

```python
year = int(input("Enter your date of birth(dob)"))
age =  2026 - year

try:
    if age <= 30 and age>=20:
        print("Age is valid!")
    else:
        raise dobException
except dobException:
    print("Sorry, your age is not valid")
```
