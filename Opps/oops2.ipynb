## Inheritance 
It allows a class to inherit attributes and methods from another class. 
(Basic operations related to Inheritance)

```python
## Inheritance (single inheritance)
## Parent class
class Car:
    def __init__(self,windows,doors,engine):
        self.windows = windows
        self.doors = doors
        self.engine = engine 
    
    def drive(self):
        print(f"The person will drive {self.engine} car")
        
```

```python
car1 = Car(4,3,"petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,engine,is_selfdriving):
        super().__init__(windows,doors,engine) ##Inheriting 
        self.is_selfdriving = is_selfdriving
        
    def selfdriving(self):
        print(f"Tesla supports self driving: {self.is_selfdriving}")
```

```python
tesla1 = Tesla(4,5,'electic',True)
tesla1.selfdriving()
tesla1.drive()
```

```python
### Multiple Inheritance
## When a class inherits from more than one base class

##Base Class 1
class Animal:
    def __init__(self,name):
        self.name = name
        
    def speak(self):
        print("Subclasses must implement this method")

##Base Class 2
class Pet:
    def __init__(self,owner):
        self.owner = owner


##Derived Class
class Dog(Animal, Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    
    def speak(self):
        return f"{self.name} says woof"

##Creating an object

dog = Dog("Buddy","Tada")
print(dog.speak())
print(f"Owner: {dog.owner}")
```
