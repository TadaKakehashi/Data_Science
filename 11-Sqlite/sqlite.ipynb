### SQL and SQlite
SQL is a standard language for managing and manipulating relational databases. 
SQLite is a self contained, serverless and zero-configuration database engine that is widely used for embedded database systems. 

```python
import sqlite3 
```

```python
### Connect to an SQLite database
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
## Create a Table
cursor.execute('''
            Create Table If Not Exists employee(
                id Integer Primary Key,
                name Text Not Null,
                age Integer,
                department Text
            )
               ''')

## Commit the changes
connection.commit()
```

```python
cursor.execute('''
            Select * from employee
               ''')
```

```python
### Inserting Data 
cursor.execute('''
            Insert Into employee (name,age,department)
                values('Tada',32,'Biker')   
            ''')
cursor.execute('''
            Insert Into employee (name,age,department)
                values('Ray',12,'Doctor')   
            ''')

connection.commit()
```

```python
### Query the data from the table
cursor.execute('Select *from employee')
rows = cursor.fetchall() #get all the records from the select * command


for row in rows:
    print(row)
```

```python
## Update the data in the table

cursor.execute(
    '''
        UPDATE employee 
        Set age = 34 
        Where name = "Ray"
    '''
)

connection.commit()
```

```python
cursor.execute('Select *from employee')
rows = cursor.fetchall() #get all the records from the select * command

for row in rows:
    print(row)
```

#### Deleting 

```python
### Deleting the data from the Table
cursor.execute(
    '''
        Delete from employee
        Where name = "Ray"
    '''
)

connection.commit()
```
