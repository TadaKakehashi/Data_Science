## File Operation


```python
### Read a Whole File
with open('example.txt','r') as file:
    content = file.read()
    print(content)
    
```

```python
### Read a File line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip(), end =" ") ### Strip removes the newline character
```

```python
### Writing a file(Overwriting)
with open('example.txt','w') as file:
    file.write('Hello World\n')
    file.write('This is a new line\n')
```

```python
### Write a file (Without Overriding)
with open('example.txt','a') as file:  # a-> append mode
    file.write("Append operation taking place!\n")
    file.write("\n")
```

```python
### Writing a list of lines to a file
line = ['First line \n','Second line \n','Third line \n']
with open('example.txt','a') as file:
    file.writelines(line)
```

```python
### Read from a source txt file and write to a destination txt file

with open('example.txt','r') as file:
    content = file.read()

with open('destination.txt','w') as dstFile:
    dstFile.write(content)
```
