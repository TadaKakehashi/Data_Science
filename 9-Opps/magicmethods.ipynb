## Magic Methods
It is also known as dunder methods (double underscore methods), are special methods that start and end with double underscores. These methods enable you to define the behavior of objects for built-in operations, such as arithmetic operations, comparisons,and more.

They are predefined methods in Python that you can override to change the behavior of your objects.

```python
class Person:
    pass

person = Person()
print(person)
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
person = Person("Tada",24)
print(person)
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

# __str__: Returns a string representation of an object
    def __str__(self): #overriding the __str__
        return f"{self.name},{self.age} years old"
    
    def __repr__(self): #Returns an official string representation of an object
        return f"Person(name={self.name},age={self.age})"
person = Person("Tada",24)
print(person)
print(repr(person))
```
